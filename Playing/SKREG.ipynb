```python
#Normalize: Normalize input data 
# Executing the constructor LinearRegression() we get the linear_regression instance.
# Model is created via the method fit(X,y)
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

n_samples=50
X=np.zeros((n_samples,2))
y=np.zeros((n_samples,2))
print(X.shape)
```

```python
X[:,0]=np.arange(X.shape[0])
y[:,0]=5+X[:,0]*np.random.rand(n_samples)*0.4
```

```python
plt.figure(figsize=(10,8))
plt.plot(X[:,0],y[:,0],'r.')
plt.xlabel('X')
plt.ylabel('Observation')
plt.grid()
```

```python
linear_regression=linear_model.LinearRegression()
#Train the model
linear_regression.fit(X,y)
w=linear_regression.coef_
b=linear_regression.intercept_
```

```python
print('Coefficient:',w, '\n Intercept:', b)
```

```python
y_hat=b[0]+w[0][0]*X[:,0]
y_hat1=linear_regression.predict(X)
```

```python
plt.figure(figsize=(10,8))
plt.plot(X[:,0],y[:,0],'r.',label='Observations')
plt.plot(X[:,0],y_hat,'b',label='Regression')
plt.plot(X[:,0],y_hat1[:,0],'k*',label='Regression_1')
plt.xlabel('X')
plt.ylabel('Observation')
plt.legend()
plt.grid()
```

```python
y[:,0]=X[:,0]**3 + np.random.rand(n_samples)*20
linear_regression.fit(X,y)
y_predicted=linear_regression.predict(X)
plt.plot(X,y_predicted,'k-')
plt.plot(X,y,'yo')
```
