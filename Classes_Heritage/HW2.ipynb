```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
class para:
  def __init__(self,t,dt,v0,theta):
    self.t=t
    self.v0=v0
    self.dt=dt
    self.theta=theta
    self.g=9.8
  #Method to calculate positions
  def positions(self):
    x=np.empty(1)
    y=np.empty(1)
    count=0
    for i in np.arange(0,self.t,self.dt):
      x1=self.v0*np.cos(self.theta)*i
      y1=self.v0*np.sin(self.theta)*i-0.5*self.g*(i**2)
      if x[count]>0 and y[count]<=0:
        break
      x=np.append(x,x1)
      y=np.append(y,y1)
      count=count+1
    return x,y
  #Method to show the plots
  def plots(self):
    x1,y1=self.positions()
    plt.figure(figsize=(10,7))
    plt.plot(x1,y1,'-r')
    plt.hlines(0,np.min(x1),np.max(x1),'g')
    plt.xlim(np.min(x1),np.max(x1))
    plt.xlabel("Range(m)")
    plt.ylabel("Height(m)")
    plt.grid()

```

```
projectil1=para(12,0.001,12,45)
```

```
x,y=projectil1.positions()
```

```
plt.plot(x,y)
```

```
projectil1.plots()
```
