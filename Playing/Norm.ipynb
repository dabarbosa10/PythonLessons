```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x=np.linspace(-6,6,200)
plt.figure(num=0,figsize=(10,8))
plt.plot(x,norm.pdf(x,loc=0,scale=1),color='black',label="$\\mu=0$, $\\sigma=1$")
plt.plot(x,norm.pdf(x,loc=1,scale=1.5),color='blue',label="$\\mu=1$, $\\sigma=1.5$")
plt.plot(x,norm.pdf(x,loc=-1,scale=0.5),color='red',label="$\\mu=-1$, $\\sigma=0.5$")
plt.title("PDF of normal distributions")
plt.ylabel("p(x)")
plt.legend()
plt.savefig('Normal.pdf')
plt.show()
```
