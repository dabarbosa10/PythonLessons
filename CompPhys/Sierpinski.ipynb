```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1=np.array([100,150,200])
y1=np.array([100,np.sqrt(200**2 - 50**2),100])
```

```python
def Sier(x,y,x0,y0,N):
    xk=np.array([x0])
    yk=np.array([y0])
    for i in range(1,N):
        n=np.random.randint(1,4)
        if n==1:
            xk1=(xk[i-1]+x[0])/2
            yk1=(yk[i-1]+y[0])/2
        if n==2:
            xk1=(xk[i-1]+x[1])/2
            yk1=(yk[i-1]+y[1])/2
        if n==3:
            xk1=(xk[i-1]+x[2])/2
            yk1=(yk[i-1]+y[2])/2
        xk=np.append(xk,xk1)
        yk=np.append(yk,yk1)
    return (xk,yk)
        
            
    
```

```python
xf,yf=Sier(x1,y1,160,120,15000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x1,y1,'k.',xf,yf,'b.')
plt.savefig('Satistical.pdf')
plt.show()
```
