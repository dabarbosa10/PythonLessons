```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn import datasets, linear_model
from sklearn.preprocessing import PolynomialFeatures

from matplotlib import pyplot as plt, rcParams
fig = plt.figure(figsize=(8, 6))
```

```python
N_train=50
sigma_train=3.5
```

```python
x=np.linspace(0,1,N_train)
```

```python
#Let's take random numbers between [0,1)
#with the N_train size
s=(sigma_train)*np.random.rand(N_train)
#propose a linear function y=2*x+s
y=2*x+s
plt.figure(figsize=(8, 6))
plt.plot(x,y,"o",ms=7, alpha=0.5, label='Training')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
x1=x.reshape(-1,1)
lin_reg.fit(x1,y)
```

```python
x2=np.linspace(0,1,70)
x2=x2.reshape(-1,1)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x,y,"o",ms=7, alpha=0.5, label='Training')
plt.plot(x,lin_reg.predict(x1),"r", alpha=0.5, label='Linear')
plt.grid()
plt.show()
```

```python
poly1=PolynomialFeatures(3)
feat1=poly1.fit_transform(x1)
```

```python
poly1.fit(feat1,y)
lin_reg2=LinearRegression()
lin_reg2.fit(feat1,y)
```

```python
Xplot=poly1.fit_transform(x1)
```

```python
Yplot=lin_reg2.predict(Xplot)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x,y,"o",ms=7, alpha=0.5, label='Training')
plt.plot(x,lin_reg.predict(x1),"r", alpha=0.5, label='Linear')
plt.plot(x,Yplot,"g", alpha=0.5, label='Poly3')
#plt.ylim(-2,4)
plt.grid()
plt.legend()
plt.show()
```

```python
poly2=PolynomialFeatures(10)
x10=poly2.fit_transform(x1)
lin_reg3=LinearRegression()
lin_reg3.fit(x10,y)
```

```python
Xplot1=poly2.fit_transform(x1)
Yplot1=lin_reg3.predict(Xplot1)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x,y,"o",ms=7, alpha=0.5, label='Training')
plt.plot(x,lin_reg.predict(x1),"r", alpha=0.5, label='Linear')
plt.plot(x,Yplot,"g", alpha=0.5, label='Poly3')
plt.plot(x,Yplot1,"b", alpha=0.5, label='Poly10')
#plt.ylim(-2,4)
plt.grid()
plt.legend()
plt.show()
```

```python
N_test=10
x_test=1.2*np.random.rand(N_test)
x_test=np.sort(x_test)
```

```python
s_test=sigma_train*(np.random.rand(N_test))
y_test=2*x_test + s_test
plt.figure(figsize=(8, 6))
plt.plot(x_test,y_test,"o",ms=7, alpha=0.5, label='Test')
#plt.ylim(-2,4)
plt.grid()
plt.legend()
plt.show()
```

```python
x1_test=x_test.reshape(-1,1)
x3P_test=poly1.fit_transform(x1_test)
x10P_test=poly2.fit_transform(x1_test)
Y1L=lin_reg.predict(x1_test)
YP3=lin_reg2.predict(x3P_test)
YP10=lin_reg3.predict(x10P_test)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_test,y_test,"o",ms=7, alpha=0.5, label='Test')
plt.plot(x_test,Y1L,"r", alpha=0.5, label='Linear')
plt.plot(x_test,lin_reg2.predict(x3P_test),"-g", alpha=0.5, label='Poly3')
plt.plot(x1_test,YP10,"b", alpha=0.5, label='Poly10')
plt.ylim(-2,4)
plt.grid()
#plt.legend()
plt.show()
```
