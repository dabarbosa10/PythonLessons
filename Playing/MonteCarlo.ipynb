```python
from random import random
```

```python
def decays(ti,tf):
    NTl=1000         #Thallium atoms
    NPb=0            #Lead atoms
    tau=3.053*60     #Half life of thallium in sec
    h=1              #Size of time-step in seconds
    p=1-2**(-h/tau)  #Prob of decay in one step
    tmax=1000.       #Total time
    tpoints=np.arange(ti,tf,h)
    Tlpoints=[]
    Pbpoints=[]
    for t in tpoints:
        Tlpoints.append(NTl)
        Pbpoints.append(NPb)
        decay=0
        for i in range(NTl):
            if random()<p:
                decay+=1
        NTl -=decay
        NPb +=decay
    return tpoints, Tlpoints,Pbpoints
    
```

```python
tin,Tlatm,Pbatm=decays(0,1000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(tin,Tlatm,'-r',label='Thallium atoms')
plt.plot(tin,Pbatm,'-b',label='Lead atoms')
plt.xlabel('Time(s)',fontsize=20)
plt.ylabel('Number of atoms',fontsize=20)
plt.grid()
plt.legend(fontsize=15)
plt.savefig('AtomsMC.pdf')
plt.show()
```
