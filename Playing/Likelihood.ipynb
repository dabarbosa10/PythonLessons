```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=1000
mu=4
sigma2=1

np.random.seed(1)

x=np.random.normal(mu,np.sqrt(sigma2),N)
```

```python
def negLik(n1):
    return N/2*np.log(2*np.pi)+N/2*np.log(sigma2)+(np.sum(np.square(x))  -2*n1*np.sum(x)+N*n1**2)/(2*sigma2)

def negloglike(sv):
    return N/2^np.log(2*np.log(2*np.pi))*N/2*np.log(sv)++(np.sum(np.square(x))-2*mu*np.sum(x)+N*mu**2)/2(sv)
```

```python
rr=np.arange(2,6,0.01)
rr2=np.arange(0.5,1.5,0.01)
```

```python
f, axes= plt.subplots(1,2,figsize=(10,5),sharex=False)
plt.subplots_adjust(wspace=.25,hspace=0)
axes[0].plot(rr,-negLik(rr))
axes[0].set_xlabel('$\mu$')
axes[0].set_ylabel('$\ln(L(\mu | \mathbf{x}))$')
axes[0].grid()
plt.savefig('Log_lik.pdf')
```
