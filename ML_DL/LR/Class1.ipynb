```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
X,y= make_blobs(n_samples=100,n_features=2, centers=2, random_state=10)
```

```python
X0=X[y==0]
X1=X[y==1]
```

```python
plt.figure(figsize=(10,8))
plt.plot(X0[:,0],X0[:,1],'og',label='Group 1')
plt.plot(X1[:,0],X1[:,1],'^b',label='Group 2')
plt.xlabel('Feature1 x(1)', fontsize=20)
plt.ylabel('Feature2 x(2)', fontsize=20)
plt.grid()
```
