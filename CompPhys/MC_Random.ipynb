```python
import numpy as np
import matplotlib.pyplot as plt
import random

%matplotlib inline
```

```python
random.seed(10)
x=np.array([0])
y=np.array([0])
```

```python
def randomW(Jmax,x1,y1):
    for i in range(1,Jmax+1):
        x1=np.append(x1,x1[i-1]+(random.random()-0.5)*2)
        y1=np.append(y1,y1[i-1]+(random.random()-0.5)*2)
    return x1, y1
```

```python
xw1,yw1=randomW(10000,x,y)
xw2,yw2=randomW(10000,x,y)
xw3,yw3=randomW(10000,x,y)
xw4,yw4=randomW(10000,x,y)
xw5,yw5=randomW(10000,x,y)
xw6,yw6=randomW(10000,x,y)
```

```python
plt.figure(figsize=(10,10))
plt.plot(xw1,yw1,'-r',xw2,yw2,'-b')
plt.plot(xw3,yw3,'-k',xw4,yw4,'-g')
plt.plot(xw5,yw5,'-y',xw6,yw6,'-m')
plt.show()
#plt.savefig('Random.pdf')
```
