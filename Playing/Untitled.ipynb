```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cuad(x):
    return x**2
```

```python
x1=np.array([-1,1/2,2])
f_x=cuad(x1)
print(f_x)
```

```python
def lag(rn,x,fx):
    x_1=x[0]
    x_2=x[1]
    x_3=x[2]
    y_1=fx[0]
    y_2=fx[1]
    y_3=fx[2]
    pl_1=y_1/((x_1-x_2)*(x_1-x_3))
    pl_2=y_2/((x_2-x_1)*(x_2-x_3))
    pl_3=y_3/((x_3-x_1)*(x_3-x_2))
    inter=np.zeros(len(rn))
    for i in range(len(rn)):
        inter[i]=((rn[i]-x_2)*(rn[i]-x_3)*pl_1)+((rn[i]-x_1)*(rn[i]-x_3)*pl_2)+((rn[i]-x_1)*(rn[i]-x_2)*pl_3)
    return (rn,inter)
```

```python
rang=np.linspace(-3,3)
xpl,lagrn=lag(rang,x1,f_x)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x1,f_x,'rd',ms=15, label='Points')
plt.plot(xpl,lagrn,'-k',label='Lagrange Interpolation')
plt.grid()
plt.legend()
plt.xlabel('x',fontsize=25)
plt.ylabel('f(x)',fontsize=25)
plt.savefig('Lag_simple.pdf')
plt.show()
```
