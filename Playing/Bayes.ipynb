```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
%matplotlib inline
```

```python
x=np.linspace(0,1,1000)
y=scipy.stats.beta.pdf(x,2,2)
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.xlim(0,1)
plt.ylim(0,1.8)
plt.ylabel(r"Beta($\theta$,2,2)",fontsize=20)
plt.xlabel(r"$ \theta $",fontsize=20)
plt.grid()
plt.savefig('Beta.pdf')
plt.show()
```

```python
def posterior(heads, tails):
    x1=np.linspace(0,1,1000)
    y1=scipy.stats.beta.pdf(x,heads+2,tails+2)
    return x1,y1
```

```python
x_1,y_1=posterior(5,10)
x_2,y_2=posterior(10,5)
x_3,y_3=posterior(5,5)
plt.figure(figsize=(10,10))
plt.plot(x_1,y_1,'r',label='5H,10T')
plt.plot(x_2,y_2,'b',label='5H,10T')
plt.plot(x_3,y_3,'k',label='5H,5T')
plt.xlim(0,1)
plt.ylim(0,5)
plt.ylabel(r"P($\theta$|D)",fontsize=20)
plt.xlabel(r"$ \theta $",fontsize=20)
plt.grid()
plt.legend(fontsize=20)
plt.savefig('Posterior.pdf')
plt.show()
```
